```python
from neuromaps import NeuroMapManuscript

model_full = NeuroMapManuscript.load("checkpoints/mu_full/model.ckpt")
model_exclude_subbox = NeuroMapManuscript.load("checkpoints/mu_exclude_subbox/model.ckpt")
model_exclude_ep = NeuroMapManuscript.load("checkpoints/mu_exclude_ep/model.ckpt")
```

```python
import json
from pathlib import Path

import matplotlib.pyplot as plt

BASE = Path("checkpoints")
RUNS = {
    "mu_full": BASE / "mu_full" / "history.json",
    "mu_exclude_subbox": BASE / "mu_exclude_subbox" / "history.json",
    "mu_exclude_ep": BASE / "mu_exclude_ep" / "history.json",
}

histories = {}
for name, path in RUNS.items():
    if path.exists():
        with open(path, "r", encoding="utf-8") as f:
            histories[name] = json.load(f)
    else:
        print(f"[skip] {name}: file not found -> {path}")

if not histories:
    raise FileNotFoundError("Не найден ни один history.json")

plt.figure(figsize=(8, 5))
for name, h in histories.items():
    if "val_loss" in h and len(h["val_loss"]) > 0:
        plt.plot(h["val_loss"], label=f"{name}: val_loss")

plt.yscale("log")
plt.xlabel("Epoch")
plt.ylabel("Validation loss (log)")
plt.title("manuscript_2: сравнение val_loss")
plt.grid(True, alpha=0.3)
plt.legend()
plt.show()

for name, h in histories.items():
    if "val_loss" in h and len(h["val_loss"]) > 0:
        print(f"{name:20s} best val_loss = {min(h['val_loss']):.6e}")
```
