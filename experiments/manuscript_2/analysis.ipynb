```python
from neuromaps import NeuroMapManuscript

model_full = NeuroMapManuscript.load("checkpoints/mu_full/model.ckpt")
model_exclude_subbox = NeuroMapManuscript.load("checkpoints/mu_exclude_subbox/model.ckpt")
```

```python
import matplotlib.pyplot as plt
import matplotlib as mpl
import numpy as np

plt.style.use('default')
mpl.rcdefaults()

hist_full = model_full.training_history
hist_exclude_subbox = model_exclude_subbox.training_history

fig, ax = plt.subplots(1, 1, figsize=(10, 4))

ax.plot(np.arange(len(hist_full["train_loss"])), hist_full["train_loss"],
        label="full train_loss", lw=1.5, alpha=.8)
ax.plot(np.arange(len(hist_full["val_loss"])), hist_full["val_loss"],
        label="full val_loss", lw=1.5, alpha=.8)
ax.plot(np.arange(len(hist_exclude_subbox["train_loss"])), hist_exclude_subbox["train_loss"],
        label="exclude subbox train_loss", lw=1.5, alpha=.8)
ax.plot(np.arange(len(hist_exclude_subbox["val_loss"])), hist_exclude_subbox["val_loss"],
        label="exclude subbox val_loss", lw=1.5, alpha=.8)

ax.set_xlabel(f"Epoch")
ax.set_ylabel("Train loss")
ax.set_ylim([0, 0.05])
ax.legend()
ax.grid(True, ls="--", alpha=.3)
plt.tight_layout()
plt.show()
```
