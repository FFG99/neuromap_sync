```python
from neuromaps import NeuroMapOriginal, NeuroMapFixed

model_original = NeuroMapOriginal.load("checkpoints/original/model.ckpt")
fixed_model = NeuroMapFixed.load("checkpoints/fixed/model.ckpt")

model_original.training_history, fixed_model.training_history
```

```python
import matplotlib.pyplot as plt
import numpy as np

hist_o = model_original.training_history
hist_f = fixed_model.training_history

plt.style.use("seaborn-v0_8")
fig, ax = plt.subplots(1, 1, figsize=(6, 4))

skip = 200
ax.plot(np.arange(skip, len(hist_o["val_loss"])), hist_o["val_loss"][skip:],
        label="Original (error) – val", lw=1.5, alpha=.8)
ax.plot(np.arange(skip, len(hist_f["val_loss"])), hist_f["val_loss"][skip:],
        label="Fixed – val", lw=1.5, alpha=.8)

ax.set_xlabel("Epoch (after 50)")
ax.set_ylabel("Loss")
ax.set_title("Training history: error-scale vs correct-scale")
ax.legend()
ax.grid(True, ls="--", alpha=.3)
plt.tight_layout()
plt.show()
```
